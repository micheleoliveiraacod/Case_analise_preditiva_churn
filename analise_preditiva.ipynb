### Case - Analise Preditiva de Churn

A partir do resultado da primeira analise, foi criado um algritimo de machine learning para rotineiramente analisar a base de clientes ativos e trazer a probabilidade destes clientes dar churn.
A partir disso a liderança pode montar um plano estrategico de acompanhamento e tratamento dos casos, afim de previnir o churn.

```python
# Passo 1: Importar e transformar a base de dados
import pandas as pd
import sklearn as sk 

tabela = pd.read_csv("dataset_cancelamentos.csv")
display(tabela)


# ver tipos de dados - corrigir os tipos de dados
display(tabela.info())
# excluir as linhas que têm valores vazios
tabela = tabela.dropna() # tabela.dropna() - excluir valores vazios / tabela.dropduplicates() - excluir linhas duplicadas
# exibir as informações da tabela após as transformações
display(tabela.info())
```

```python
# Passo 2 - Preparar a base de dados para o algoritimo de machine learning
# A coluna alvo é a coluna que queremos prever (cancelou)
# As outras colunas são as colunas de entrada (features)
# Verificar o tipo de dado de cada coluna
# Transformar os dados textuais em dados numéricos (features)

# duracao_cliente -> de texto para número
# assinatura -> de texto para número
# cancelou -> de numero para texto

display(tabela.info())

# int -> numero inteiro
# float -> numero com casa decimal
# object -> texto

# Importar LabelEncoder - transforma dados textuais em dados numéricos, com base na frequência e variação dos dados da coluna
from sklearn.preprocessing import LabelEncoder

# duracao_cliente
# Annual - 1
# Monthly - 2
# Quarterly - 3

codificador_duracao = LabelEncoder()
tabela["duracao_contrato"] = codificador_duracao.fit_transform(tabela["duracao_contrato"])
display(tabela.info())

# assinatura
# Basic - 1
# Standard - 2
# Premium - 3
codificador_assinatura = LabelEncoder()
tabela["assinatura"] = codificador_assinatura.fit_transform(tabela["assinatura"])
display(tabela.info())
```

```python
# A coluna cancelou precia ser transformada de número para texto, pois é o que quero prever
# Criar Nova Coluna (Recomendado - Preserva Original)

# cancelou
# Não - 0 (ativo)
# Sim - 1 (risco de cancelamento)

tabela['cancelou_texto'] = tabela['cancelou'].map({0: 'ativo', 1: 'risco de cancelamento'})
display(tabela.info())
```

```python
# Preparar os dados de treino e teste, informando o que é x e o que é y
# X coluna que eu quero usar para prever
# Y colunas que eu usar para treinar o modelo, pois trazem os fatos reais sobre o cancelamento

# y -> é a coluna da base de dados que eu quero prever
y = tabela["cancelou_texto"]

# x -> as colunas da base de dados que eu vou usar para prever o y
X = tabela[[ "duracao_contrato", "ligacoes_callcenter", "dias_atraso", "assinatura", "idade" ]]

# separar em dados de treino e dados de teste
from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.3) # 30% dos dados para teste, 70% para treino
```

```python
# Boas praticas de codigo, salvar as colunas do modelo em uma variavel
colunas_modelo = x_treino.columns
```

```python
# Passo 3 - Criar e treinar o algoritmo de machine learning com os 70% dos dados (dados de treino)
# Criar o modelo: Cancelou -> 0 - ativo, 1 - risco de cancelamento

# Arvore de Decisão -> RandomForest
# Nearest Neighbors -> KNN -> Vizinhos Próximos

# 1. importar o algoritimo
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

# 2. criar o algoritimo
modelo_arvoredecisao = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()

# 3. treinar o algoritimo
modelo_arvoredecisao.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)
```

```python
# Passo 4 - Escolher qual o melhor modelo fazendo previsões com os dois modelos, com os 30% dos dados que separei para teste
# Fazer previsões com os dois modelos
previsao_arvoredecisao = modelo_arvoredecisao.predict(x_teste)
previsao_knn = modelo_knn.predict(x_teste)

# Calular a acurácia de cada um dos modelos
from sklearn.metrics import accuracy_score
display(accuracy_score(y_teste, previsao_arvoredecisao))
display(accuracy_score(y_teste, previsao_knn))

# O melhor modelo foi o de Arvore de Decisão (RandomForest)
# Neste caso, a acurácia foi de aproximadamente 95% e o modeloesta aprovado para fazer as previsões reais
```

```python
# Passo 5 - Usar o melhor modelo para fazer previsão de novos clientes
# melhor modelo é o modelo_arvoredecisao

# importar os novos clientes para fazer a previsao
tabela_novos_clientes = pd.read_csv("dataset_novos_clientes.csv")
ids_clientes = tabela_novos_clientes["c"].copy()


# ajustar os dados dos novos clientes para o formato do modelo
# duracao_cliente
codificador_duracao = LabelEncoder()
tabela_novos_clientes["duracao_contrato"] = codificador_duracao.fit_transform(tabela_novos_clientes["duracao_contrato"])
display(tabela_novos_clientes.info())
# assinatura
codificador_assinatura = LabelEncoder()
tabela_novos_clientes["assinatura"] = codificador_assinatura.fit_transform(tabela_novos_clientes["assinatura"])
display(tabela_novos_clientes.info())
# cancelou_texto
tabela_novos_clientes['cancelou_texto'] = tabela_novos_clientes['cancelou'].map({0: 'ativo', 1: 'risco de cancelamento'})
tabela_novos_clientes = tabela_novos_clientes[colunas_modelo]
display(tabela_novos_clientes.info())
```

```python
# Fazer a previsão com o melhor modelo
nova_previsao = modelo_arvoredecisao.predict(
    tabela_novos_clientes[colunas_modelo]
)
display(nova_previsao)

# Adicionar a previsão na tabela de novos clientes
tabela_novos_clientes["previsao"] = nova_previsao
display(tabela_novos_clientes)
```

```python
# Resultado de todos os clientes com risco de cancelamento, para enviar para o time que vai fazer a campanha de retenção

resultado = pd.DataFrame({
    "CustomerID": ids_clientes,
    "previsao": nova_previsao
})

display(resultado)
```
